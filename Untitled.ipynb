```python
import numpy as np
import torch
import os
from matplotlib import pyplot as plt
```

```python
dataset = 'paintings'
```

```python
vanilla_path = f"results/vanilla/{dataset}/acc.npy"
vanilla_acc = np.load(vanilla_path)
vanilla_acc = vanilla_acc.mean()
```

```python
finetune_path = f"results/finetune/{dataset}/acc.npy"
finetune_acc = np.load(finetune_path)
finetune_acc = finetune_acc.mean()
```

```python
flow_folder = "results/flow"
flow_setting_list = os.listdir(flow_folder)
print(flow_setting_list)
```

```python
plt.axhline(y=vanilla_acc, label="vanilla")
plt.axhline(y=finetune_acc, label="finetune")
for i, exp in enumerate(flow_setting_list):
    file_path = os.path.join(flow_folder, exp, f'{dataset}/acc.npy')
    flow_acc = np.load(file_path)
    flow_acc[1], flow_acc[6] = flow_acc[6].copy(), flow_acc[1].copy()
    flow_acc[2], flow_acc[7] = flow_acc[7].copy(), flow_acc[2].copy()
    flow_acc_list = []
    x = [0, 1, 5, 10, 20, 30, 40, 50]
    for j in range(flow_acc.shape[1]):
        flow_step_acc = flow_acc[:, j]
        flow_step_mean = flow_step_acc.mean()
        flow_acc_list.append(flow_step_mean)
    plt.plot(x, flow_acc_list, label=exp)
plt.title(dataset)
plt.legend()
plt.show()
```
